```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
import xgboost as xgb
from statsmodels.tsa.arima.model import ARIMA
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras import backend as K
import warnings
warnings.filterwarnings("ignore")


def evaluate_model(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r = max(y_true) - min(y_true)
    accuracy = 1 - (rmse / r)
    print(f"{model_name} -> MAE: {mae:.2f}, RMSE: {rmse:.2f}, Accuracy: {accuracy:.2%}")
    return mae, rmse


def plot_forecast(y_true, y_pred, title):
    plt.figure(figsize=(10, 4))
    plt.plot(y_true, label='Actual', marker='o')
    plt.plot(y_pred, label='Predicted', marker='x')
    plt.title(f"{title} - Actual vs Forecasted")
    plt.xlabel("Time")
    plt.ylabel("Demand")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


def process_individual_dataset(df, name="Dataset"):
    print(f"\n{'='*30}\n Processing {name}\n{'='*30}")
    df = df.copy()
    df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
    df = df.sort_values('ORDERDATE')

    y = df['QUANTITYORDERED'].values
    X_time = np.arange(len(y)).reshape(-1, 1)

    # Train-Test Split
    split_idx = int(0.8 * len(y))
    y_train, y_test = y[:split_idx], y[split_idx:]
    X_train, X_test = X_time[:split_idx], X_time[split_idx:]

    metrics = {}

    # ----- LSTM -----
    scaler = MinMaxScaler()
    y_scaled = scaler.fit_transform(y.reshape(-1, 1)).flatten()

    def create_lstm_sequences(data, lookback=10):
        X, y = [], []
        for i in range(len(data) - lookback):
            X.append(data[i:i+lookback])
            y.append(data[i+lookback])
        return np.array(X), np.array(y)

    lookback = 10
    X_lstm_all, y_lstm_all = create_lstm_sequences(y_scaled, lookback)
    split_lstm = int(0.8 * len(X_lstm_all))
    X_lstm_train, X_lstm_test = X_lstm_all[:split_lstm], X_lstm_all[split_lstm:]
    y_lstm_train, y_lstm_test = y_lstm_all[:split_lstm], y_lstm_all[split_lstm:]

    X_lstm_train = X_lstm_train.reshape((X_lstm_train.shape[0], lookback, 1))
    X_lstm_test = X_lstm_test.reshape((X_lstm_test.shape[0], lookback, 1))

    K.clear_session()
    model_lstm = Sequential([
        LSTM(64, input_shape=(lookback, 1)),
        Dense(1)
    ])
    model_lstm.compile(optimizer='adam', loss='mse')
    model_lstm.fit(X_lstm_train, y_lstm_train, epochs=10, batch_size=32, verbose=0)

    lstm_pred = model_lstm.predict(X_lstm_test).flatten()
    lstm_pred = scaler.inverse_transform(lstm_pred.reshape(-1, 1)).flatten()
    y_lstm_actual = scaler.inverse_transform(y_lstm_test.reshape(-1, 1)).flatten()
    metrics['LSTM'] = evaluate_model(y_lstm_actual, lstm_pred, 'LSTM')
    plot_forecast(y_lstm_actual, lstm_pred, f"{name} - LSTM")

    # ----- XGBoost -----
    xgb_model = xgb.XGBRegressor()
    xgb_model.fit(X_train, y_train)
    xgb_pred = xgb_model.predict(X_test)
    metrics['XGBoost'] = evaluate_model(y_test, xgb_pred, 'XGBoost')
    plot_forecast(y_test, xgb_pred, f"{name} - XGBoost")

    # ----- ARIMA -----
    arima_model = ARIMA(y_train, order=(5, 1, 0))
    arima_fit = arima_model.fit()
    arima_pred = arima_fit.forecast(steps=len(y_test))
    metrics['ARIMA'] = evaluate_model(y_test, arima_pred, 'ARIMA')
    plot_forecast(y_test, arima_pred, f"{name} - ARIMA")

    # ----- SVR -----
    svr_model = SVR()
    svr_model.fit(X_train, y_train)
    svr_pred = svr_model.predict(X_test)
    metrics['SVR'] = evaluate_model(y_test, svr_pred, 'SVR')
    plot_forecast(y_test, svr_pred, f"{name} - SVR")

    # ----- Random Forest -----
    rf_model = RandomForestRegressor(random_state=42)
    rf_model.fit(X_train, y_train)
    rf_pred = rf_model.predict(X_test)
    metrics['Random Forest'] = evaluate_model(y_test, rf_pred, 'Random Forest')
    plot_forecast(y_test, rf_pred, f"{name} - Random Forest")

    # ----- Gradient Boosting -----
    gb_model = GradientBoostingRegressor()
    gb_model.fit(X_train, y_train)
    gb_pred = gb_model.predict(X_test)
    metrics['Gradient Boosting'] = evaluate_model(y_test, gb_pred, 'Gradient Boosting')
    plot_forecast(y_test, gb_pred, f"{name} - Gradient Boosting")

    # ----- Linear Regression -----
    lr_model = LinearRegression()
    lr_model.fit(X_train, y_train)
    lr_pred = lr_model.predict(X_test)
    metrics['Linear Regression'] = evaluate_model(y_test, lr_pred, 'Linear Regression')
    plot_forecast(y_test, lr_pred, f"{name} - Linear Regression")

    return metrics


# Example Usage
# Sample dataset (replace this with your own)
df_sample = pd.read_csv("sales_data_sample.csv", encoding="cp1252")
df_sample['ORDERDATE'] = pd.to_datetime(df_sample['ORDERDATE'])
df_sample = df_sample[df_sample['PRODUCTLINE'] == 'Classic Cars']
process_individual_dataset(df_sample, "Classic Cars")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from scipy.stats import poisson
import xgboost as xgb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras import backend as K
import seaborn as sns

def process_individual_dataset(df, name="Dataset"):
    print(f"\n{'='*30}\n Processing {name}\n{'='*30}")

    df = df.copy()
    df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
    df = df.sort_values(['ORDERDATE'])

    # Label encode DEALSIZE
    deal_mapping = {'Small': 1, 'Medium': 2, 'Large': 3}
    df['DEALSIZE_ENC'] = df['DEALSIZE_ENC'].map(deal_mapping) if df['DEALSIZE_ENC'].dtype == 'O' else df['DEALSIZE_ENC']

    # Markov Transition Matrix
    def build_markov_transition(df, col='DEALSIZE_ENC'):
        states = df[col].tolist()
        matrix = pd.DataFrame(0, index=[1, 2, 3], columns=[1, 2, 3])
        for i, j in zip(states[:-1], states[1:]):
            matrix.loc[i, j] += 1
        matrix = matrix.div(matrix.sum(axis=1), axis=0).fillna(0)
        return matrix

    markov_matrix = build_markov_transition(df)

    def get_markov_probs(state):
        row = markov_matrix.loc[state]
        return pd.Series({
            'P_STAY': row[state],
            'P_GROW': row[state+1] if state < 3 else 0,
            'P_SHRINK': row[state-1] if state > 1 else 0
        })

    markov_probs = df['DEALSIZE_ENC'].apply(get_markov_probs)
    df = pd.concat([df.reset_index(drop=True), markov_probs.reset_index(drop=True)], axis=1)

    # Poisson Process
    lambda_est = df['QUANTITYORDERED'].mean()
    df['POISSON_LAMBDA'] = lambda_est
    df['POISSON_SIM'] = poisson.rvs(mu=lambda_est, size=len(df))

    # LSTM
    def create_lstm_sequences(data, lookback):
        X, y = [], []
        for i in range(len(data) - lookback):
            X.append(data[i:i+lookback])
            y.append(data[i+lookback])
        return np.array(X), np.array(y)

    scaler = MinMaxScaler()
    qty_scaled = scaler.fit_transform(df[['QUANTITYORDERED']])
    
    best_mae = float('inf')
    best_lookback = None
    best_lstm_preds = None
    best_X_lstm = None
    best_y_lstm = None

    for lookback in [5, 10, 15, 20]:
        K.clear_session()
        X_lstm, y_lstm = create_lstm_sequences(qty_scaled.flatten(), lookback)
        if len(X_lstm) == 0: continue
        X_lstm = X_lstm.reshape((X_lstm.shape[0], X_lstm.shape[1], 1))

        model = Sequential([
            LSTM(64, input_shape=(lookback, 1)),
            Dense(1)
        ])
        model.compile(optimizer='adam', loss='mse')
        model.fit(X_lstm, y_lstm, epochs=5, batch_size=32, verbose=0)

        preds = model.predict(X_lstm, verbose=0)
        mae = mean_absolute_error(y_lstm, preds)

        if mae < best_mae:
            best_mae = mae
            best_lookback = lookback
            best_lstm_preds = preds
            best_X_lstm = X_lstm
            best_y_lstm = y_lstm

    lstm_preds = best_lstm_preds
    X_lstm = best_X_lstm
    y_lstm = best_y_lstm

    # XGBoost
    feature_cols = ['PRICEEACH', 'MSRP', 'DEALSIZE_ENC', 'POISSON_LAMBDA', 'POISSON_SIM', 'P_STAY', 'P_GROW', 'P_SHRINK']
    df = df.dropna(subset=feature_cols + ['QUANTITYORDERED'])

    xgb_model = xgb.XGBRegressor()
    xgb_model.fit(df[feature_cols], df['QUANTITYORDERED'])
    xgb_preds = xgb_model.predict(df[feature_cols])
    xgb_trimmed = xgb_preds[-len(lstm_preds):]

    # Meta Model
    meta_X = pd.DataFrame({
        'lstm_pred': lstm_preds.flatten(),
        'xgb_pred': xgb_trimmed
    })
    meta_y = df['QUANTITYORDERED'].values[-len(meta_X):]

    X_train, X_test, y_train, y_test = train_test_split(meta_X, meta_y, test_size=0.2, random_state=42)
    meta_model = RandomForestRegressor(random_state=42)
    meta_model.fit(X_train, y_train)
    final_preds = meta_model.predict(X_test)

    # Evaluation
    mae = mean_absolute_error(y_test, final_preds)
    rmse = np.sqrt(mean_squared_error(y_test, final_preds))
    r = max(y_test) - min(y_test)
    accuracy = 1 - (rmse / r)


    print(f"Markov Transition Matrix:\n{markov_matrix}")
    print(f"Poisson λ: {lambda_est:.2f} | First 5 Simulated: {df['POISSON_SIM'].head().tolist()}")
    print(f"Best LSTM Lookback: {best_lookback} | LSTM MAE: {best_mae:.2f}")
    print(f"Meta-Model MAE: {mae:.2f}")
    print(f"Meta-Model RMSE: {rmse:.2f}")
    print(f"Meta-Model Accuracy: {accuracy:.2%}")

    # PLOT: Actual vs Forecasted
    plt.figure(figsize=(10, 5))
    sorted_idx = np.argsort(y_test) 
    plt.plot(np.arange(len(y_test)), y_test[sorted_idx], label='Actual', marker='o')
    plt.plot(np.arange(len(final_preds)), final_preds[sorted_idx], label='Predicted', marker='x')
    plt.title(f"Actual vs Forecasted Demand - {name}")
    plt.xlabel("Sample Index")
    plt.ylabel("Quantity Ordered")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
#--------------------
    
    
    errors = np.array(y_test) - np.array(final_preds)
    
    plt.figure(figsize=(8, 5))
    sns.histplot(errors, kde=True, bins=20, color='coral')
    plt.title("Error Distribution (Actual - Forecasted)")
    plt.xlabel("Error")
    plt.ylabel("Frequency")
    plt.axvline(x=0, color='black', linestyle='--', linewidth=1)
    plt.tight_layout()
    plt.savefig("error_distribution.png", dpi=300)
    plt.show()
#---------------------------------------
# --- Loop Over All Product Lines ---
datasets = {
    #"Motorcycles": d_mt,
    "Classic Cars": d_cars,
    #"Trains": d_trains,
    #"Trucks & Buses": d_tb
}

for name, df in datasets.items():
    process_individual_dataset(df, name)
```
